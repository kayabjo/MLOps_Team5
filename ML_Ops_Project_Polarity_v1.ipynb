## <font color='blue'>Load json files & combine</font>

```python
oct20 = pd.read_json('CNN/articles/cnn_scrape-2020-10.jsonl', lines=True)
nov20 = pd.read_json('CNN/articles/cnn_scrape-2020-11.jsonl', lines=True)
oct20.shape, nov20.shape
```

```python
data = pd.concat([oct20,nov20], ignore_index=True)
data.shape
```

## <font color='blue'>Create Polarity & Subjectivity Columns</font>

```python
%time data['polarity'] = data['body'].apply(lambda x: TextBlob(x).polarity)
```

```python
%time data['subjectivity'] = data['body'].apply(lambda x: TextBlob(x).subjectivity)
```

```python
data.head()
```

## <font color='blue'> Create Month & Year columns</font>

```python
data['year']  = data['published_date'].apply(lambda x: dt.strptime(x,'%Y-%m-%d').year)
data['month'] = data['published_date'].apply(lambda x: dt.strptime(x,'%Y-%m-%d').month)
```

```python
data.head()
```

```python
data.describe()
```

## <font color='blue'>Save Updated File</font>

```python
data.to_json('CNN/cnn_oct20-nov20.jsonl', orient='records', lines=True)
```

## <font color='blue'> PIVOT Values by Category </font>

```python
agg_func_p_math = {
    'polarity':
    ['mean','median','min','max','std','var']
}
data.groupby(['site_category']).agg(agg_func_p_math).round(4)
```

```python
agg_func_s_math = {
    'subjectivity':
    ['mean','median','min','max','std','var']
}
data.groupby(['site_category']).agg(agg_func_s_math).round(4)
```

```python
data.month[data['month']==11].count()
```

## Comparison between Oct20 & Nov20

```python
agg_func_p_math = {
    'polarity':
    ['mean','median','min','max','std','var']
}

data[(data['month']==10)|(data['month']==11)].groupby(['month','site_category']).agg(agg_func_p_math).round(4)
```

```python
agg_func_s_math = {
    'subjectivity':
    ['mean','median','min','max','std','var']
}

data[(data['month']==10)|(data['month']==11)].groupby(['month','site_category']).agg(agg_func_s_math).round(4)
```
